```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
import senescyt
```

```python
DIR_SYMBOLS = './symbols_4'
DIR_TEST = './img_test'
```

```python
st = senescyt.symbols(DIR_SYMBOLS)
```

```python
# Read test images
images = {}
for ifile in os.listdir(DIR_TEST):
    if ifile.endswith(".png"):
        path = os.path.join(DIR_TEST, ifile)
        label, _ = os.path.splitext(ifile)
        img = cv2.imread(path)
        images[label] = img
```

```python
idxs = list(images.keys())
image_idx = None
```

```python
@interact(idx=idxs)
def show_img(idx):
    global image_idx
    print(f"Label = {idx}")
    print(f"Dim = {images[idx].shape[1]} x {images[idx].shape[0]}")
    plt.imshow(images[idx][:,:,::-1])
    plt.show()
    image_idx = idx
```

```python
if image_idx is not None:
    img_bw = senescyt.img2bw(images[image_idx], threshold=senescyt.THR)
    detections = senescyt.decode(img_bw, st)
    # Draw detection box
    images_with_detections = []
    for detection in detections:
        i_ = images[image_idx].copy()
        cv2.rectangle(
            i_,
            (detection["TOP_LEFT_X"], detection["TOP_LEFT_Y"]),
            (detection["BOTTOM_RIGHT_X"], detection["BOTTOM_RIGHT_Y"]),
            detection["COLOR"],
            1
        )
        images_with_detections.append(i_)
    # Plot detections
    fig, axs = plt.subplots(nrows=len(images_with_detections), ncols=1, figsize=(16,12))
    for idx, ax in enumerate(axs):
        ax.imshow(images_with_detections[idx][:,:,::-1])
        ax.set_title(f"Label = {detections[idx]['LABEL']}, Match = {detections[idx]['MATCH_VALUE']}")

    plt.subplots_adjust(left=0.1,
                        bottom=0.1, 
                        right=0.9, 
                        top=0.9, 
                        wspace=0.5, 
                        hspace=0.5)
    plt.show()
    string_decoded = senescyt.get_string(detections)
    print(f"String Decoded = {string_decoded}")
```
